### Data Analysis

## DATA QUALITY CHECK

```python
print('\n=== DATA QUALITY ASSESSMENT ===')

#Check for missing values
print('\n Missing Values:\n')
missing_data = df.isnull().sum()
print(missing_data[missing_data > 0])

#Check for Duplicates
print(f'\n Duplicate Rows : {df.duplicated().sum()}')
```

## Data Type Conversion

```python
# Convert data Columns to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%Y-%m-%d')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%Y-%m-%d')

#Check Conversion
print("\nDate columns converted:\n")
print(f"Order Date dtype: {df['Order Date'].dtype}")
print(f"Ship Date dtype: {df['Ship Date'].dtype}")
```

## UNIVARIATE ANALYSIS

```python
# Univariate Analysis is the simplest form of staristical analysis.As the name suggests uni
#means "one" and "variate" means "variable"- it is the analysis of a singnle variable.

print('\n === CATEGORICAL VARIABLES ===\n')

categorical_cols= ['Customer Segment','Category','Region','payment Mode','Delivery Status','Auto Reorder']

for col in categorical_cols:
    print(f'\n---{col} ---')
    print(df[col].value_counts())

    # Visualisation
    plt.figure(figsize=(6,6))
    df[col].value_counts().plot(kind='bar',color='skyblue')
    plt.title(f'Distibution of {col}')
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
```

## Numerical Variables Analysis

```python
print('\n === NUMERIVAL VARIABLES ===')

numerical_cols = ['Quantity','Unit Price','Discount (%)', 'Sales Amount', 'Cost Price','Profit','Stock Left','Reorder Quantity']

for col in numerical_cols:
    print(f'\n---{col} ---\n')
    print(f'Mean:{df[col].mean():.2f}')
    print(f'Median:{df[col].median():.2f}')
    print(f'Std Dev:{df[col].std():.2f}')
    print(f'Min:{df[col].min():.2f}')
    print(f'Max:{df[col].max():.2f}')

    #Visuralization
    plt.figure(figsize=(8,4))

    plt.subplot(1, 2, 1)
    df[col].hist(bins=30,color='lightgreen',alpha=0.7)
    plt.title(f'Distribution of {col}')

    plt.subplot(1, 2, 2)
    df.boxplot(column=col)
    plt.title(f'Boxplot of {col}')

    plt.tight_layout()
    plt.show()
```

## BIVARIATE ANALYSIS

```python
# Bivariaten Analysis is the statistical of two variables at the same time. The core purose of the investigate relationship between them.

print('\n=== SALES & PROFIT BY CATEGORY ===\n')

category_performance = df.groupby('Category').agg({
    'Sales Amount':'sum',
    'Profit': 'sum',
    'Order ID':'count'
}).rename (columns={'Order ID':'Order Count'})

category_performance ['Profit Margin'] =(category_performance ['Profit']/category_performance['Sales Amount']) *100
print(category_performance,'\n\n')

#Visualization
fig, axes=plt.subplots(2,2, figsize=(15,12))

#Sales by Category
category_performance['Sales Amount'].plot(kind='bar',ax=axes[0,1],color='green',alpha=0.7)
axes[0,0].set_title('Total Sales by Category')
axes[0,1].tick_params(axis='x',rotation=30)

#Profit by Category
category_performance['Profit'].plot(kind='bar', ax=axes[0,0],color='blue',alpha=0.7)
axes[0,1].set_title('Total Profit by Category')
axes[0,1].tick_params(axis='x',rotation=30)

#Profit Margin
category_performance['Profit Margin'].plot(kind='bar',ax=axes[1,0],color='Orange',alpha=0.7)
axes[1,0].set_title('Profit Margin by Category (%)')
axes[1,0].tick_params(axis='x',rotation=30)

#Order Count
category_performance['Order Count'].plot(kind='bar', ax=axes[1,1],color='red',alpha=0.7)
axes[1,1].set_title('Order Count by Category')
axes[1,1].tick_params(axis='x',rotation=30)

plt.tight_layout()
plt.show()
```

## Regional Performance

```python
print('\n=== REGIONAL PERFORMANCE ===\n')

# Aggregation
regional_performance = df.groupby('Region').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'Order ID': 'count'
}).rename(columns={'Order ID': 'Order Count'})

# Profit Margin
regional_performance['Profit Margin'] = (
    regional_performance['Profit'] / regional_performance['Sales Amount']
) * 100

print(regional_performance, '\n')

# ================= Visualization =================
plt.figure(figsize=(12, 8))

# Sales Distribution
plt.subplot(2, 2, 1)
regional_performance['Sales Amount'].plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Sales Distribution by Region')
plt.ylabel('')

# Profit Distribution
plt.subplot(2, 2, 2)
regional_performance['Profit'].plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Profit Distribution by Region')
plt.ylabel('')

# Order Count
plt.subplot(2, 2, 3)
sns.barplot(
    x=regional_performance.index,
    y=regional_performance['Order Count']
)
plt.title('Order Count by Region')
plt.xticks(rotation=30)

# Profit Margin
plt.subplot(2, 2, 4)
sns.barplot(
    x=regional_performance.index,
    y=regional_performance['Profit Margin']
)
plt.title('Profit Margin by Region (%)')
plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

## TIME SERIES ANALYSIS

```python
print('\n=== TIME SERIES ANALYSIS ===\n')

# Ensure datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract time components
df['Order Month'] = df['Order Date'].dt.to_period('M')
df['Order Year'] = df['Order Date'].dt.year

# Monthly trends
monthly_trends = (
    df.groupby('Order Month')
      .agg({
          'Sales Amount': 'sum',
          'Profit': 'sum',
          'Order ID': 'count'
      })
      .rename(columns={'Order ID': 'Order Count'})
)

# Convert PeriodIndex to Timestamp for plotting
monthly_trends.index = monthly_trends.index.to_timestamp()

print('Monthly Trends:')
print(monthly_trends.tail(10), '\n')

# ================= Visualization =================
plt.figure(figsize=(15, 10))

# Monthly Sales
plt.subplot(3, 1, 1)
monthly_trends['Sales Amount'].plot(marker='o')
plt.title('Monthly Sales Trend')
plt.ylabel('Sales Amount')
plt.grid(True)

# Monthly Profit
plt.subplot(3, 1, 2)
monthly_trends['Profit'].plot(marker='o')
plt.title('Monthly Profit Trend')
plt.ylabel('Profit')
plt.grid(True)

# Monthly Order Count
plt.subplot(3, 1, 3)
monthly_trends['Order Count'].plot(marker='o')
plt.title('Monthly Order Count Trend')
plt.ylabel('Order Count')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## ADVANCEED ANALYSIS

```python
print('\n=== CORRELATION ANALYSIS ===\n')

# Select numerical columns for correlation
numerical_df=df[['Quantity', 'Unit Price','Discount (%)','Sales Amount', 'Cost Price','Profit']]

# Correlation matrix
correlation_matrix = numerical_df.corr()
print('Correlation Matrix :')
print(correlation_matrix, '\n\n')

# Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True,linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

# Customer Segment Analysis

```python
print('\n ==== CUSTOMER SEGMENT ANALYSIS === \n')

segment_analysis =  df.groupby('Customer Segment').agg({
    'Sales Amount': ['sum','mean'],
    'Profit': ['sum','mean'],
    'Order ID': 'count',
    'Discount (%)': 'mean'
}).round(2)

segment_analysis.columns = ['Total Sales', 'Avg Sales', 'Total Profit', 'Avg Profit', 'Order Count', 'Avg Discount']
print(segment_analysis, '\n\n')

#Visualization

fig,axes = plt.subplots(2,2, figsize=(10,15))

segment_analysis['Total Sales'].plot(kind='bar',ax=axes[0,0], color='purple')
axes[0,1].set_title('Total Sales by Customer Segment')
axes[0,0].tick_params(axis='x',rotation=30)

segment_analysis['Total Profit'].plot(kind='bar',ax=axes[0,1], color='red')
axes[0,1].set_title('Total Profit by Customer Segment')
axes[0,0].tick_params(axis='x',rotation=30)

segment_analysis['Avg Profit'].plot(kind='bar',ax=axes[1,0], color='blue')
axes[0,1].set_title('Avg Profit by Customer Segment')
axes[0,0].tick_params(axis='x',rotation=30)

segment_analysis['Avg Discount'].plot(kind='bar',ax=axes[1,1], color='yellow')
axes[0,1].set_title('Avg Discount by Customer Segment')
axes[0,0].tick_params(axis='x',rotation=30)

plt.tight_layout()
plt.show()
```

# Key Insight & Summary

```python
print('\n === KEY INSIGHTS === \n')

#Overall Matrix
total_sales = df['Sales Amount'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
avg_prioft_margin = (total_profit / total_sales) * 100

print(f'OVERALL PERFORMANCE:\n')
print(f'Total Sales: ${total_sales:,.2f}')
print(f'Total Profit: ${total_profit:,.2f}')
print(f'Total Orders: {total_orders:,}')
print(f'Average Profit Margin: {avg_prioft_margin:.2f}%')

#Top Performing Categories
top_categories = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print(f'\nTOP PERFORMING CATEGORIES:\n')
for i,(catogry,profit) in enumerate(top_categories.items(),1):
  print(f'{i}. {catogry}: ${profit:,.2f}')

#Regional Performance
best_region = df.groupby('Region')["Profit"].sum().idxmax()
print(f'\n BEST PERFORMANCE REGION: {best_region}')

# Delivery Performance
delivery_status = df['Delivery Status'].value_counts(normalize=True)*100
print(f'\n DELIVEY STATUS:')
for status,percentage in delivery_status.items():
  print(f'{status}:{percentage:.1f}%')

# Customer Segments
best_segment = df.groupby('Customer Segment')['Profit'].sum().idxmax()
print(f'\n MOST PROFITABLE CUSTOMER SEGMENT:{best_segment}')
```

# EXPORT RESULTS

```python
analysis_results ={
    'total_sales':total_sales,
    'total_profit':total_profit,
    'total_orders':total_orders,
    'avg_profit_margin':avg_prioft_margin,
    'top_categories':top_categories.index[0],
    'best_region':best_region,
    'best_customer_segment':best_segment

}

# Convert to DataFrame and save
results_df = pd.DataFrame(analysis_results, index=[0])
results_df.to_csv('eda_analysis_sumary.csv',index=False)
print("\n Analysis summary saved to 'eda_analysis_summary.csv'")
```
